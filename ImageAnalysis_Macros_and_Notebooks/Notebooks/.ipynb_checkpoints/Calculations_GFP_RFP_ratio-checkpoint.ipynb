# Analysis GFP/RFP ratio at heart inflowtract

This jupyter notebook can be used to analyze line measurements of the fluorescence in two channels at 6 z-planes in 3 embryos.

Measurements included in GitHub "../Data/Fluorescence_intensity_ratio/line_measure".


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io 
import os
import seaborn as sns
```

```python
# list files with measurements from directory
path="../Data/Fluorescence_intensity_ratio/line_measure/"
list_dir=os.listdir(path)
```

```python
# read all files into a DataFrame
df=pd.DataFrame([])

for i in list_dir:
    if ".csv" in i: 
        df_read=pd.read_csv(path+i)
        df_read["Slice"]=int(i.split("_z")[1].replace(".csv",""))
        df_read["Position"]=i.split("_")[0]
        df_read["Channel"]=i.split("_")[1]
        df=df.append(df_read)
        
df_crop =df[df["Distance_(microns)"]<60]

df_crop.head()
```

```python
# Subset data for embryo 1 --> Pos1
df_p1=df_crop[df_crop.Position=="Pos1"].copy()

# normalize by max intensity for GFP and RFP
df_p1_GFP=df_p1[df_p1.Channel=="GFP"].copy()
df_p1_GFP["Gray_Value_Norm"]=df_p1_GFP.Gray_Value/np.max(df_p1_GFP.Gray_Value)

df_p1_RFP=df_p1[df_p1.Channel=="RFP"].copy()
df_p1_RFP["Gray_Value_Norm"]=df_p1_RFP.Gray_Value/np.max(df_p1_RFP.Gray_Value).copy()
# plot before normalization
sns.lineplot(x=df_p1_GFP.columns[0],y="Gray_Value",data=df_p1_GFP,color="g")
sns.lineplot(x=df_p1_GFP.columns[0],y="Gray_Value",data=df_p1_RFP,color="r")
#plt.savefig("./Embryo1_Gray_values_GFP_RFP.pdf")
```

```python
# plot after normalization
sns.lineplot(x=df_p1_GFP.columns[0],y="Gray_Value_Norm",data=df_p1_GFP,color="g")
sns.lineplot(x=df_p1_GFP.columns[0],y="Gray_Value_Norm",data=df_p1_RFP,color="r")
#plt.savefig("D:/Ines/measurements_myl7-FliV/Embryo1_Gray_values_minmax-norm_GFP_RFP.pdf")
```

```python
# summarize embryo 1
df_p1_GFP=df_p1_GFP.rename(columns={"Gray_Value":"Embyro_1 GFP","Gray_Value_Norm":"Embyro_1 GFP_Norm"})
df_p1_RFP=df_p1_RFP.rename(columns={"Gray_Value":"Embyro_1 RFP","Gray_Value_Norm":"Embyro_1 RFP_Norm"})
df_p1_summary=pd.concat([df_p1_GFP["Distance_(microns)"],df_p1_GFP[["Embyro_1 GFP","Embyro_1 GFP_Norm"]],df_p1_RFP[["Embyro_1 RFP","Embyro_1 RFP_Norm"]]],axis=1).set_index("Distance_(microns)")


p1_GFP_mean=df_p1_GFP.groupby(df_p1_GFP.columns[0]).mean()
p1_RFP_mean=df_p1_RFP.groupby(df_p1_RFP.columns[0]).mean()
p1_ratio=(p1_GFP_mean["Embyro_1 GFP"]/p1_RFP_mean["Embyro_1 RFP"])/np.max(p1_GFP_mean["Embyro_1 GFP"]/p1_RFP_mean["Embyro_1 RFP"])
df_p1_summary.head()
#plt.plot(p1_ratio)
```

```python
# subset embryo 2
df_p2=df_crop[df_crop.Position=="Pos2"].copy()

df_p2_GFP=df_p2[df_p2.Channel=="GFP"].copy()
df_p2_GFP["Gray_Value_Norm"]=df_p2_GFP.Gray_Value/np.max(df_p2_GFP.Gray_Value)

df_p2_RFP=df_p2[df_p2.Channel=="RFP"].copy()
df_p2_RFP["Gray_Value_Norm"]=df_p2_RFP.Gray_Value/np.max(df_p2_RFP.Gray_Value)
#sns.lineplot(x=df_p2_GFP.columns[0],y="Gray_Value",data=df_p2_GFP,color="g")
#sns.lineplot(x=df_p2_GFP.columns[0],y="Gray_Value",data=df_p2_RFP,color="r")
#plt.savefig("./Embryo2_Gray_values_GFP_RFP.pdf")
```

```python
# summarize embryo 2
df_p2_GFP=df_p2_GFP.rename(columns={"Gray_Value":"Embyro_2 GFP","Gray_Value_Norm":"Embyro_2 GFP_Norm"})
df_p2_RFP=df_p2_RFP.rename(columns={"Gray_Value":"Embyro_2 RFP","Gray_Value_Norm":"Embyro_2 RFP_Norm"})
df_p2_summary=pd.concat([df_p2_GFP["Distance_(microns)"],df_p2_GFP[["Embyro_2 GFP","Embyro_2 GFP_Norm"]],df_p2_RFP[["Embyro_2 RFP","Embyro_2 RFP_Norm"]]],axis=1).set_index("Distance_(microns)")

p2_GFP_mean=df_p2_GFP.groupby(df_p2_GFP.columns[0]).mean()
p2_RFP_mean=df_p2_RFP.groupby(df_p2_RFP.columns[0]).mean()
p2_ratio=(p2_GFP_mean["Embyro_2 GFP"]/p2_RFP_mean["Embyro_2 RFP"])/np.max(p2_GFP_mean["Embyro_2 GFP"]/p2_RFP_mean["Embyro_2 RFP"])
```

```python
# subset embryo 4
df_p4=df_crop[df_crop.Position=="Pos4"].copy()

df_p4_GFP=df_p4[df_p2.Channel=="GFP"].copy()
df_p4_GFP["Gray_Value_Norm"]=df_p4_GFP.Gray_Value/np.max(df_p4_GFP.Gray_Value)

df_p4_RFP=df_p4[df_p4.Channel=="RFP"].copy()
df_p4_RFP["Gray_Value_Norm"]=df_p4_RFP.Gray_Value/np.max(df_p4_RFP.Gray_Value)
#sns.lineplot(x=df_p4_GFP.columns[0],y="Gray_Value",data=df_p4_GFP,color="g")
#sns.lineplot(x=df_p4_GFP.columns[0],y="Gray_Value",data=df_p4_RFP,color="r")
#plt.savefig("D:/Ines/measurements_myl7-FliV/Embryo4_Gray_values_GFP_RFP.pdf")
```

```python
# summarize embryo 4
df_p4_GFP=df_p4_GFP.rename(columns={"Gray_Value":"Embyro_4 GFP","Gray_Value_Norm":"Embyro_4 GFP_Norm"})
df_p4_RFP=df_p4_RFP.rename(columns={"Gray_Value":"Embyro_4 RFP","Gray_Value_Norm":"Embyro_4 RFP_Norm"})
df_p4_summary=pd.concat([df_p4_GFP["Distance_(microns)"],df_p4_GFP[["Embyro_4 GFP","Embyro_4 GFP_Norm"]],df_p4_RFP[["Embyro_4 RFP","Embyro_4 RFP_Norm"]]],axis=1).set_index("Distance_(microns)")

p4_GFP_mean=df_p4_GFP.groupby(df_p4_GFP.columns[0]).mean()
p4_RFP_mean=df_p4_RFP.groupby(df_p4_RFP.columns[0]).mean()
p4_ratio=(p4_GFP_mean["Embyro_4 GFP"]/p4_RFP_mean["Embyro_4 RFP"])/np.max(p4_GFP_mean["Embyro_4 GFP"]/p4_RFP_mean["Embyro_4 RFP"])
```

```python
# summarize all embryos
df_summary=pd.concat([df_p1_summary,df_p2_summary,df_p4_summary],axis=1)
df_summary.head()
pALL_mean=(p1_ratio+p2_ratio+p4_ratio)/3
```

```python
# Concatenate and assign final names
df_summary_ratio_short=pd.concat([p1_ratio,p2_ratio,p4_ratio,pALL_mean],axis=1)
df_summary_ratio_short.columns=["Embryo_1 ratio GFP/RFP","Embryo_2 ratio GFP/RFP","Embryo_3 ratio GFP/RFP","Mean ratio GFP/RFP"]
df_summary_ratio_short.head()
```

```python
# Example plot
# Statistics and final graphs where generated in GraphPad Prism
styleOrder=["Mean ratio GFP/RFP","Embryo_1 ratio GFP/RFP","Embryo_2 ratio GFP/RFP","Embryo_3 ratio GFP/RFP"]
sns.lineplot(data=df_summary_ratio_short,palette=["gray","gray","gray","r"],style_order=styleOrder)

#plt.savefig("./Ratios_minmax-norm_GFP_RFP.pdf")
```
