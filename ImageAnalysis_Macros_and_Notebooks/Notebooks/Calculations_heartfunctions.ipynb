# Analyze_heartfunction

This jupyter notebook can be used to analyze the results from the Fiji plugin 
Analyze_heartfunction.py.

Measurements included in GitHub "../Data/Heart_functions/Heart_function_measurements.xlsx".

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# access result table and make dataframe 
df=pd.read_excel('../Data/Heart_functions/Heart_function_measurements.xlsx',delimiter=',')
df=df.set_index('ImgTitle')
pd.options.mode.chained_assignment = None
```

```python
# define function to calculate heart functions
def Calc_HeartFun(df_OE2d):
    df_OE2d['Volume_diastole_Vent']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_diastole_Ventricle_1']*(df_OE2d.loc[:,'Short_ax_diastole_Ventricle_1'])**2);
    df_OE2d['Volume_diastole_Vent_2']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_diastole_Ventricle_2']*(df_OE2d.loc[:,'Short_ax_diastole_Ventricle_2'])**2);
    df_OE2d['Volume_diastole_Vent_Mean_um3']=((df_OE2d['Volume_diastole_Vent']+ df_OE2d['Volume_diastole_Vent_2'])/2);
    df_OE2d['Volume_systole_Vent']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_systole_Ventricle_1']*(df_OE2d.loc[:,'Short_ax_systole_Ventricle_1'])**2);
    df_OE2d['Volume_systole_Vent_2']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_systole_Ventricle_2']*(df_OE2d.loc[:,'Short_ax_systole_Ventricle_2'])**2);
    df_OE2d['Volume_systole_Vent_Mean_um3']=(df_OE2d['Volume_systole_Vent']+ df_OE2d['Volume_systole_Vent_2'])/2;
    df_OE2d['Stroke_Volume_Ventricle_nl']=((df_OE2d.Volume_diastole_Vent_Mean_um3)-(df_OE2d.Volume_systole_Vent_Mean_um3))*(10**-9)*1000;
    df_OE2d['Ejection_fraction_Ventricle_%']=100*(((df_OE2d.Volume_diastole_Vent_Mean_um3)-(df_OE2d.Volume_systole_Vent_Mean_um3))/(df_OE2d.Volume_diastole_Vent_Mean_um3));
    df_OE2d['Cardiac_Output_Ventricle_nl/min']=((df_OE2d.Stroke_Volume_Ventricle_nl)*(df_OE2d.HR_Bps*60));
    df_OE2d['Volume_diastole_Atrium']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_diastole_Atrium_1']*(df_OE2d.loc[:,'Short_ax_diastole_Atrium_1'])**2)
    df_OE2d['Volume_diastole_Atrium_2']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_diastole_Atrium_2']*(df_OE2d.loc[:,'Short_ax_diastole_Atrium_2'])**2)
    df_OE2d['Volume_diastole_Atrium_Mean_um3']=((df_OE2d['Volume_diastole_Atrium']+ df_OE2d['Volume_diastole_Atrium_2'])/2)
    df_OE2d['Volume_systole_Atrium']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_systole_Atrium_1']*(df_OE2d.loc[:,'Short_ax_systole_Atrium_1'])**2)
    df_OE2d['Volume_systole_Atrium_2']=((1/6)*math.pi)*(df_OE2d.loc[:,'Long_ax_systole_Atrium_2']*(df_OE2d.loc[:,'Short_ax_systole_Atrium_2'])**2)
    df_OE2d['Volume_systole_Atrium_Mean_um3']=(df_OE2d['Volume_systole_Atrium']+ df_OE2d['Volume_systole_Atrium_2'])/2
    df_OE2d['Stroke_Volume_Atrium_nl']=((df_OE2d.Volume_systole_Atrium_Mean_um3)-(df_OE2d.Volume_diastole_Atrium_Mean_um3))*(10**-9)*1000
    df_OE2d['Ejection_fraction_Atrium_%']=100*(((df_OE2d.Volume_systole_Atrium_Mean_um3)-(df_OE2d.Volume_diastole_Atrium_Mean_um3))/(df_OE2d.Volume_systole_Atrium_Mean_um3))
    df_OE2d['Cardiac_Output_Atrium_nl/min']=((df_OE2d.Stroke_Volume_Atrium_nl)*(df_OE2d.HR_Bps*60))
    SV_AT=df_OE2d['Stroke_Volume_Atrium_nl']
    EF_AT=df_OE2d['Ejection_fraction_Atrium_%']
    CO_AT=df_OE2d['Cardiac_Output_Atrium_nl/min']
    SV_V=df_OE2d['Stroke_Volume_Ventricle_nl']
    EF_V=df_OE2d['Ejection_fraction_Ventricle_%']
    CO_V=df_OE2d['Cardiac_Output_Ventricle_nl/min']
    HR=(df_OE2d.HR_Bps)
    DVolV=df_OE2d['Volume_diastole_Vent_Mean_um3']
    DVolA=df_OE2d['Volume_systole_Atrium_Mean_um3']
    return  [SV_AT,EF_AT,CO_AT,SV_V,EF_V,CO_V,HR,DVolV,DVolA]
```

```python
# apply function to previously defined dataframe
Results=Calc_HeartFun(df)
df_Results=pd.DataFrame(Results)
# transform the dataframe
df_Results_T=df_Results.T
df_Results_T
```

# Edit next to your filenames

```python
# reset the index and make groups according to the filename 

df_Results_T=df_Results_T.reset_index()
df_Results_T['groups']=df_Results_T.ImgTitle.str.replace('20161206-', '').str.split('-0', expand=True)[0]
```

```python
#make dataframe according to the group name
df_c2=df_Results_T[df_Results_T['groups']=="48hpf-UVR"]
df_o2=df_Results_T[df_Results_T['groups']=="48hpf-KTS"]
df_c5=df_Results_T[df_Results_T['groups']=="5dpf-UVR"]
df_o5=df_Results_T[df_Results_T['groups']=="5dpf-KTS"]

sign_all=np.empty(0)
names_all=np.empty(0)
pval_all=np.empty(0)

# make statistical test, kruskal wallis as normality is not assumed
for title_measure in df_Results_T.columns:
    if title_measure != "ImgTitle" and title_measure != "groups":
        [t1,p1]=stats.kruskal(df_c2[title_measure],df_o2[title_measure])
        [t2,p2]=stats.kruskal(df_c5[title_measure],df_o5[title_measure])
        if p1 < 0.05 and p1 > 0.01: 
            sign1="*"
        elif p1 < 0.01 and p1 > 0.001:
            sign1="**"
        elif p1 < 0.001:
            sign1="***"
        else:
            sign1=""
        if p2 < 0.05 and p2 > 0.01: 
            sign2="*"
        elif p2 < 0.01 and p2 > 0.001:
            sign2="**"
        elif p2 < 0.001:
            sign2="***"
        else:
            sign2=""
        pval=[p1,p2]
        signs=[sign1,sign2]
        names=[title_measure + "_2d_Control-vs-OE",title_measure + "_5d_Control-vs-OE"]
        sign_all=np.append(sign_all,signs)
        pval_all=np.append(pval_all,pval)
        names_all=np.append(names_all,names)
        #else:
        #    sign_all[title_measure]=[sign1,sign2]
        #print("2d_Control-vs-OE",title_measure,"     ",p1, sign1)
        #print("5d_Control-vs-OE",title_measure,"     ",p2, sign2)
significance=pd.DataFrame()
significance["Names"]=names_all
significance["P-values"]=pval_all
significance["Significant?"]=sign_all
significance.sort_values(by=["Names"])
```
